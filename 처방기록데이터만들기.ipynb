```python
import pandas as pd
import numpy as np
prescription_record = pd.DataFrame({"patientID":[],"hosID":[],"drugID":[],"doctorID":[],"prescription_amount":[],"day":[],"state":[]})
prescription_record
```

```python
append = pd.read_csv("/content/sample_data/append.csv")
real = append.iloc[0:len(append)+1,1:3]
```

```python
num = 0
for i in drug_max.keys():
  for j in range(1,101):
    doctorID = np.random.randint(2,33)
    prescription_amount = np.random.randint(1,drug_max[i])
    prescription_record.loc[num] = [j,real.iloc[doctorID-2,1],drug_ID[i],doctorID,prescription_amount,1,1]
    num += 1
prescription_record
```

```python
prescription_record.to_csv("/content/sample_data/prescription_record.csv")
```
